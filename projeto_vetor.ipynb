```python
from sentence_transformers import SentenceTransformer
import numpy as np

#Carregar modelo de embeddings
modelo = SentenceTransformer('all-MiniLM-L6-v2')

#"Banco de dados" (frases)
dados = [
    "Eu gosto de tecnologia",
    "Adoro esportes e academia",
    "Comprei um notebook novo",
    "Futebol é meu esporte favorito"
]

#Gerar embeddings dos dados
vetores = modelo.encode(dados)

#Função de busca
def buscar(frase):
    consulta = modelo.encode([frase])[0]

    melhor = None
    maior_sim = -1

    for texto, vetor in zip(dados, vetores):
        sim = np.dot(consulta, vetor) / (np.linalg.norm(consulta) * np.linalg.norm(vetor))

        if sim > maior_sim:
            maior_sim = sim
            melhor = texto

    return melhor, maior_sim

#Teste
consulta = input("Digite algo: ")
resultado, similaridade = buscar(consulta)

print("Resultado:", resultado)
print("Similaridade:", similaridade)
```
